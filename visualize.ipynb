```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
DATASET_PATH = "TreatedRoomSmallSet/Empty"
```

```python
RIRs = np.load(os.path.join(DATASET_PATH, "deconvolved.npy"), mmap_mode='r')
print("Shape of RIRs:")
print(RIRs.shape)
```

```python
sample_RIR = RIRs[0,0,].copy()
print("Shape of sampled RIRs:")
print(sample_RIR.shape)
```

```python
time_ax = np.linspace(0, len(RIRs[0,0,:9600]) / 48000, num=len(RIRs[0,0,:9600]))
plt.plot(time_ax, sample_RIR[:9600])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

```python
sample_RIR2 = RIRs[0,1,].copy()
print("Shape of sampled RIRs:")
print(sample_RIR2.shape)
```

```python
plt.plot(time_ax, sample_RIR2[:9600])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

```python
fig, axs = plt.subplots(2, 5, figsize=(36, 12))

time_ax = np.linspace(0, len(RIRs[0,0,:4800]) / 48000, num=len(RIRs[0,0,:4800]))

axs[0, 0].plot(time_ax, RIRs[0,0,:4800])
axs[0, 0].set_title('RIR 1')
axs[0, 1].plot(time_ax, RIRs[0,1,:4800], 'tab:orange')
axs[0, 1].set_title('RIR 2')
axs[0, 2].plot(time_ax, RIRs[0,2,:4800], 'tab:green')
axs[0, 2].set_title('RIR 3')
axs[0, 3].plot(time_ax, RIRs[0,3,:4800], 'tab:red')
axs[0, 3].set_title('RIR 4')
axs[0, 4].plot(time_ax, RIRs[0,4,:4800])
axs[0, 4].set_title('RIR 5')
axs[1, 0].plot(time_ax, RIRs[0,5,:4800], 'tab:orange')
axs[1, 0].set_title('RIR 6')
axs[1, 1].plot(time_ax, RIRs[0,6,:4800], 'tab:green')
axs[1, 1].set_title('RIR 7')
axs[1, 2].plot(time_ax, RIRs[0,7,:4800], 'tab:red')
axs[1, 2].set_title('RIR 8')
axs[1, 3].plot(time_ax, RIRs[0,8,:4800], 'tab:green')
axs[1, 3].set_title('RIR 9')
axs[1, 4].plot(time_ax, RIRs[0,9,:4800], 'tab:red')
axs[1, 4].set_title('RIR 10')



# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
fig.savefig("output_rirs.png")
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(24, 12))

time_ax = np.linspace(0, len(RIRs[0,0,:4800]) / 48000, num=len(RIRs[0,0,:4800]))

axs[0].plot(time_ax, RIRs[0,0,:4800])
axs[0].set_title('RIR 1')
axs[1].plot(time_ax, RIRs[1,0,:4800], 'tab:orange')
axs[1].set_title('RIR 2')
axs[2].plot(time_ax, RIRs[2,0,:4800], 'tab:green')
axs[2].set_title('RIR 3')



# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
fig.savefig("output_rirs_human.png")
plt.show()
```
